```python
import pandas as pd

import warnings
```

```python
warnings.filterwarnings('ignore')
```

```python
areas = pd.read_csv('state-areas.csv')
```

```python
abbr  = pd.read_csv('state-abbrevs.csv')
abbr.head()
```

```python
pop   = pd.read_csv('state-population.csv')
pop.shape
```

```python
# pop 與abbrevs合併，分別依據state/region和 abbreviation合併
#pop2 =  pd.merge(pop,abbr,left_on='state/region',right_on='abbreviation') #如果對不齊，資料會流失
#pop2.shape

pop2 =  pd.merge(pop,abbr,how='outer',left_on='state/region',right_on='abbreviation') #outer外合併，數據部會丟失，但會存在空值
pop2.head()
```

```python
pop2.drop(labels='abbreviation',axis=1,inplace=True) #inplace在原本的基礎上替代
pop2.head()
```

```python
pop2.isnull().any()
```

```python
cond = pop2['state'].isnull()

a = pop2['state/region'][cond]

#去重

a.unique()
#USA - United States
#PR  - Puerto Rico
```

```python
#查看缺失數據

cond = pop2['state/region'] =='PR'

pop2['state'][cond] ='Puerto Rico'
```

```python
#查看缺失數據

cond = pop2['state/region'] =='USA'

pop2['state'][cond] ='United States'
```

```python
# 檢查空值資料
pop2.isnull().any()
```

```python
cond = pop2['population'].isnull()
pop2[cond]
```

```python
##無法找到PR的populationg數據 乾脆刪除
pop2.dropna(inplace=True) # 把空值去除
pop2.isnull().any()
```

```python
pop2.head()
```

```python
areas.head()
```

```python
#合併 pop2 和areas

pop3 = pd.merge(pop2,areas,how='outer')

display(pop3.head,pop3.shape)
```

```python
pop3.isnull().any()
```

```python
cond = pop3['area (sq. mi)'].isnull()

pop3[cond]
```

```python
areas['area (sq. mi)'].sum() # 全美面積
```

```python
pop3.fillna(3790399,inplace=True)
```

```python
pop3.notnull().all()
```

```python
#找出2010年的全民人口數據

pop3.head()
```

```python
# 參考 pop_2010 = pop3.query("(year  ==2010 or ...) and ages == 'total'")

pop_2010 = pop3.query("(year  ==2010) and ages == 'total'")
```

```python
pop_2010.set_index('state/region',inplace=True) # 設定索引
```

```python
pop_2010
```

```python
#計算人口密度

pop_density = pop_2010['population']/pop_2010['area (sq. mi)']
pop_density
```

```python
#轉成DataFrame

pop_density = pd.DataFrame(pop_density,columns=['pop_density'])
```

```python
#全局設置
pd.set_option('display.float_format',lambda x:'%0.3f'%(x))
#pop_density
#pop_density.round(3)
pop_density
```

```python

pop_2010.head(5)
```

```python
pop_density.head(5)
```

```python
#合併
result = pop_2010.merge(pop_density,left_index=True,right_index=True)
```

```python
#排序
result.sort_values(by='pop_density',ascending=False,inplace=True) # ascending 升續

result.to_csv('./pop_data.csv')
```
