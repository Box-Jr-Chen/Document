```python
# SNS DEMO
# 結果:好友密度少、日記密度少、沒有真實頭像為假帳號。

#A: 日誌密度

#B: 好友密度

#C: 是否有頭像

#帳號比例 no:yes = 0.3:0.7
```

```python
import numpy as np
```

```python
#熵是所有機率的總和
info_D = -(0.3*np.log2(0.3)+ 0.7*np.log2(0.7))
info_D
```

```python
#gini

gini_D = 0.3*(1-0.3) + 0.7*(1-0.7)
gini_D
```

```python
#-----------------------------------
```

```python
#一個系統熵最小值是0
#如果所有值都是YES 100% 1.0
1*np.log2(1)
```

```python
# 使用日誌密度進行裂分  屬性名A
#日誌密度 
#s 0.3  -----> 2no , 1yes
#m 0.4  -----> 1no , 3yes
#l 0.3  -----> 3yes
#info_A_D = -0.3*(s的熵)-  0.4*(m的熵)-0.3*(l的熵)

info_A_D = -0.3*(2/3*np.log2(2/3) + 1/3*np.log2(1/3)) - 0.4*(1/4*np.log2(1/4) + 3/4*np.log2(3/4)) - 0.3*(1*np.log2(1))
info_A_D
```

```python
#熵減小了，系統有序了

#信息增益
info_D - info_A_D
```

```python
# C4.5使用百分比 
# 信息增益/info_B_D
0.2812908992306927 /0.6
```

```python
#gini_A_D
gini_A_D = 0.3*(2/3*(1-2/3)  + 1/3*(1-1/3)) +    0.4*(2*1/4*(1-1/4))   +    0.3*(1*(1-1))
gini_A_D
#gini數越小越純
```

```python
#-----------------------------------
```

```python
# 使用好友密度進行裂分  屬性名B
#日誌密度 
#s 0.4  -----> 3no , 1yes
#m 0.4  -----> 4yes
#l 0.2  -----> 2yes
#info_B_D = -0.4*(s的熵)-  0.4*(m的熵)-0.2*(l的熵)

info_B_D = -0.4*(3/4*np.log2(3/4) + 1/4*np.log2(1/4)) - 0.4*(1*np.log2(1)) - 0.2*(1*np.log2(1))
info_B_D
```

```python
#gini_B_D
gini_B_D = 0.4*(3/4*(1-3/4)*2)
gini_B_D
#gini數越小越純
```

```python
#熵減小了，系統有序了

#使用好友進行劃分，信息增益大
info_D - info_B_D
```

```python
# C4.5使用百分比 
# 信息增益/info_B_D
0.5567796494470396 /0.32451124978365314
```

```python
import numpy as np
from sklearn.tree import DecisionTreeClassifier

import sklearn.datasets as datasets
```

```python
iris = datasets.load_iris()
X = iris['data']

y = iris['target']
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
# classifier  
# default = gini
clf = DecisionTreeClassifier()

clf.fit(X_train,y_train)
```

```python
clf.score(X_test,y_test)
```

```python
#決策樹繪圖
import graphviz
from sklearn import tree
```

```python
data = tree.export_graphviz(clf,out_file=None,feature_names=iris.feature_names,class_names=iris.target_names,filled=True)


graph= graphviz.Source(data)
graph.render('./iris')
```

```python
#https://www.youtube.com/watch?v=m0ozxwDey5w&list=PLwDQt7s1o9J65aQ6HYN4LaqRkX8ncKG1D&index=84
# 20:13
```

```python
# classifier  
# default = gini
clf = DecisionTreeClassifier(criterion='entropy',max_depth =3)

clf.fit(X_train,y_train)
clf.score(X_test,y_test)


tree.export_graphviz(clf,out_file='tree.dot',feature_names=iris.feature_names,class_names=iris.target_names,filled=True)
```

```python
with open('./tree.dot') as fp:
    dot =fp.read()
```

```python
#隨機森林
from sklearn.ensemble import RandomForestClassifier
```

```python
forest = RandomForestClassifier(n_estimators=100) # 樹越多 花的時間越長  越複雜效果才比較明顯

forest.fit(X_train,y_train)

forest.score(X_train,y_train)
```

```python
#極度隨機森林
from sklearn.ensemble import ExtraTreesClassifier
```

```python
extraTree = ExtraTreesClassifier(n_estimators=100) # 樹越多 花的時間越長  越複雜效果才比較明顯

extraTree.fit(X_train,y_train)

extraTree.score(X_train,y_train)
```
